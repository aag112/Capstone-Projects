#create an analysis of the office supplier dataset and visualize it through 3 different charts 

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as plt
from matplotlib import pyplot as plt
sns.set_style("whitegrid")
%matplotlib inline
```

```
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/OfficeSupplies.csv')
```

```
df.head()
```

```
df.info()
```

```
print(('The last order date was on: ') + df['OrderDate'].max())
```

```
df['Units_Sales'] = df['Units'] * df['Unit Price']
df
```

```
sns.lineplot(data=df,x='Rep', y='Units_Sales')
plt.xticks(rotation=90)
plt.title('Unit Sales per Rep')
sns.set(rc={'figure.figsize':(14,8)})
```

```
df.dropna(inplace=True)
```

```
sns.barplot(x='Rep', y='Units_Sales', hue = 'Region', data=df)
```

```
plt.figure(figsize=(15,8))

sns.set(style="darkgrid")

#ax = sns.barplot(x="budget2018", y="ministere", data=budget, label="Total")
sns.set_color_codes("pastel")
sns.barplot(x="Units_Sales", y="Region", data=df, label="Sales per Region")

sns.despine(left=True, bottom=True)
plt.title("Sales per Region")
plt.tight_layout()
plt.show()
```

```
sns.histplot(data=df,x=df['Item'], hue=df['Rep'], multiple='stack');
```

I want the values inside the barplot. I must create a for loop through the patches to annotate the bars

```
g=sns.barplot(data= df,x=df['Rep'],y=df['Units_Sales'],label="Sales")
plt.title('Sales by Rep With Values Displayed')
plt.xticks(rotation=90)
plt.figure(figsize=(12,8))
ax=g
for p in ax.patches:
             ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
_ = g.set_ylim(0,2000) #To make space for the annotations
```
